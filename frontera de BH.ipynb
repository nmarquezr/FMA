```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
# Coordenadas de los dos puntos

x1=4
y1=3
x2=8
y2=6
punto1 = (x1, y1)
punto2 = (x2, y2)


# Coordenadas del centro de la circunferencia
h = (((x1)**2)-((x2)**2)+((y1)**2)-((y2)**2))/(2*(x1-x2))

# Calcular el radio de la circunferencia (distancia entre el centro y uno de los puntos)
r = np.sqrt((x2 - h)**2 + y2**2)
    
# Calcular el ángulo de inicio y final para el arco

angulo_inicio = np.arctan2(y1 , x1-h)
angulo_final = np.arctan2(y2 , x2-h)
# Crear un ángulo de paso (esto determina la suavidad del arco)
angulo_paso = 1

# Calcular los puntos en el arco
angulos = np.arange(angulo_inicio, angulo_final, angulo_paso)
x_arco = h + r * np.cos(angulos)
y_arco = r * np.sin(angulos)

angulo_inicio_grados = np.degrees(angulo_inicio)
angulo_final_grados = np.degrees(angulo_final)

fig, ax = plt.subplots()

arco2 = patches.Arc((h, 0), 2 * r, 2 * r, angle=0, theta1=0, theta2=180, linestyle='dashed', color='blue')
arco = patches.Arc((h, 0), 2 * r, 2 * r, angle=0, theta1=angulo_final_grados, theta2=angulo_inicio_grados)

# Agregar el arco a los ejes
ax.add_patch(arco2)

# Agregar el arco a los ejes
ax.add_patch(arco)


# Graficar los puntos y el arco
plt.scatter([x1, x2], [y1, y2], color='red', label='Puntos')
#plt.plot(x_arco, y_arco, label='Arco', color='blue')
plt.axis('equal')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()

```
